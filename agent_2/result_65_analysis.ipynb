```python
import json

file_path = '/home/jaesung/pulmuone/agent_2/result_data.json'

with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
import re
import pandas as pd

df_list = []

for i in range(len(data)):

    Q = data[i]['query']

    if not data[i]['response'] or data[i]['response'] == None:
        pass
    else:
        A_origin = data[i]['response'].split('A:')[-1].strip()

        # '배합비' 부분을 추출하는 정규 표현식 (재료명과 % 형식)
        matches = re.findall(r'[가-힣a-zA-Z0-9]+:\s*\d+%', A_origin)
        # matches = re.findall(r'[가-힣a-zA-Z0-9]+:\s*\d+%|[가-힣a-zA-Z0-9]+\s*\d+%', A_origin)

        A = ', '.join(matches)

        # print(Q)
        # print('')
        # print(A)
        # print('')

        len_ox = 0
        for q in Q.split(', '):
            if q.strip() in A:
                len_ox += 1

        # print(len_ox)        

        # 정규 표현식을 사용해 ':' 뒤에 있는 숫자들을 추출 (숫자 + % 형식)
        numbers = re.findall(r':\s*(\d+)%', A)

        # 추출된 숫자들을 정수로 변환 후 합산
        total_sum = sum(map(int, numbers))

        # print('')
        # print(f"숫자의 합: {total_sum}")
        
        # 데이터프레임에 추가할 항목 생성
        df_list.append({
            'Q': Q,
            'A': A,
            'len_ox': len_ox,
            'total_sum': total_sum
        })
    
df = pd.DataFrame(df_list)
```

```python
import re

def adjust_percentages(row):
    percentages = re.findall(r'(\d+)%', row['A'])
    percentages = list(map(int, percentages))
    
    total = sum(percentages)
    
    if total > 0:
        adjusted_percentages = [round(p * 100 / total, 2) for p in percentages]
    else:
        adjusted_percentages = percentages

    new_a = re.sub(r'(\d+)%', '{}%', row['A']).format(*adjusted_percentages)
    
    row['A'] = new_a
    row['total_sum'] = sum(adjusted_percentages)
    
    return row

df = df.apply(adjust_percentages, axis=1)
```
